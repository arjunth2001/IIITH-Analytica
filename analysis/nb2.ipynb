```python
import pandas as pd
import json
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
with open("princeton/data_analysis.json") as f:
    data = json.load(f)
```

```python
years = []
guidelines= []
perc = []

for y in data.keys():
    for g in data[y].keys():
        p = data[y][g]
        years.append(int(y))
        guidelines.append(g)
        perc.append(p)
```

```python
df = pd.DataFrame({"Year":years,"FTC Guideline":guidelines,"Percentage":perc})
```

```python
fig = plt.figure(figsize=(15,8))
fig.patch.set_facecolor("white")
plt.title("Longitudinal Analysis of FTC Guideline")
plt.suptitle("Percentage of FTC Guidelines Appearing in Privacy Policies over the Years from 2000 to 2019 in the Princeton Longitudinal Dataset as predicted by our models", fontsize=15, ha='center')
plt.xticks(years)
sns.lineplot(data=df, x="Year", y="Percentage", hue="FTC Guideline",style="FTC Guideline")
plt.legend(bbox_to_anchor=(1,1))
plt.savefig("images/long.png",dpi=300,bbox_inches="tight")
plt.show()
```
