```python
import pandas as pd
```

```python
df = pd.read_csv("./all.csv")
```

```python
df2 = pd.read_csv("./annotate_lookup.csv")
```

```python
df = df[["id", "data", "label", "site_id"]].dropna()
```

```python
df
```

```python
df2
```

```python
df3 = pd.merge(df, df2, on="site_id")
df3 = df3.drop_duplicates("id")
df3
```

```python
row_list = []
```

```python
for indx, row in df3.iterrows():
    for label in row["label"].split("#"):
        row_list.append({
            "context": row["policy_text"],
            "question": label,
            "answer": row["data"],
        })
```

```python
df4 = pd.DataFrame(row_list)
```

```python
df4
```

```python
df4.to_csv("evals1.csv")
```
