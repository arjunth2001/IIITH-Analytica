```python
import pandas as pd
```

```python
df = pd.read_csv("PrivacyQA_EMNLP.csv")
```

```python
df = df[["text","context","answer_start","question"]]
```

```python
df
```

```python
df2 = pd.read_csv("policyQA.csv")
```

```python
df2
```

```python
df2.columns=["context","text","question","answer_start"]
df2
```

```python
result = pd.concat([df,df2])
result=result.reset_index(drop = True)
```

```python
result
```

```python
train=result.sample(frac=0.8,random_state=200) #random state is a seed value
test=result.drop(train.index)
```

```python
train = train.reset_index(drop = True)
test= test.reset_index(drop = True)
```

```python
train.to_csv("../datasets/QnA/train.csv",index=False)
```

```python
test.to_csv("../datasets/QnA/val.csv",index=False)
```
