```python
import pandas as pd
```

```python
df = pd.read_csv("opp115.csv")
```

```python
df = df[["data_practice","text"]]
```

```python
df["data_practice"].value_counts()
```

```python
df2 = pd.read_csv("PrivacyQA_EMNLP/data/meta-annotations/OPP-115 Annotations/test_opp_annotations.csv",sep="\t")
```

```python
df2
```

```python
def process(df):
    df["data_practice"]= ["Other"]*len(df)
    for i in range(len(df)):
        try:
            if df["first"][i]==1:
                df["data_practice"][i]= "First Party Collection/Use"

            if df["third"][i]==1:
                df["data_practice"][i]= "Third Party Sharing/Collection"

            if df["dataretention"][i]==1:
                df["data_practice"][i]= "Data Retention"

            if df["datasecurity"][i]==1:
                df["data_practice"][i]= "Data Security"

            if df["user_access"][i]==1:
                df["data_practice"][i]= "User Access, Edit and Deletion"

            if df["user_choice"][i]==1:
                df["data_practice"][i]= "User Choice/Control"

            if df["audiences"][i]==1:
                df["data_practice"][i]= "International and Specific Audiences"
        except:
            pass
    return df
```

```python
df2 = process(df2)
df2["data_practice"].value_counts()
df2 = df2[["Query","data_practice"]]
df2.columns=["text","data_practice"]
```

```python
df3 = pd.read_csv("PrivacyQA_EMNLP/data/meta-annotations/OPP-115 Annotations/train_opp_annotations.csv",sep="\t")
```

```python
df3 = process(df3)
df3["data_practice"].value_counts()
df3 = df3[["Query","data_practice"]]
df3.columns=["text","data_practice"]
```

```python
result = pd.concat([df,df2,df3])
result=result.reset_index(drop = True)
```

```python
result
```

```python
train=result.sample(frac=0.8,random_state=200) #random state is a seed value
test=result.drop(train.index)
```

```python
train = train.reset_index(drop = True)
test= test.reset_index(drop = True)
```

```python
train.to_csv("../datasets/OPP/train.csv",index=False)
```

```python
test.to_csv("../datasets/OPP/val.csv",index=False)
```
